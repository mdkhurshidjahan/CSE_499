```python
def count_images_in_folder(folder_path):
    image_count = 0
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp')):
                image_count += 1
    return image_count

def visualize_image_counts(dataset_folder):
    folder_counts = {}
    for dataset_type in ['train', 'test']:
        dataset_type_path = os.path.join(dataset_folder, dataset_type)
        for root, dirs, _ in os.walk(dataset_type_path):
            for folder in dirs:
                folder_path = os.path.join(root, folder)
                image_count = count_images_in_folder(folder_path)
                folder_name = os.path.join(dataset_type, folder)
                if folder_name not in folder_counts:
                    folder_counts[folder_name] = image_count
                else:
                    folder_counts[folder_name] += image_count
    
    for folder, image_count in folder_counts.items():
        print(f"Folder: {folder}\t| Number of Images: {image_count}")
    
    folder_names = list(folder_counts.keys())
    image_counts = list(folder_counts.values())

    # Generate random colors for the bars
    num_bars = len(folder_names)
    colors = plt.cm.Set3(np.linspace(0, 1, num_bars))

    plt.figure(figsize=(10, 6))
    plt.bar(folder_names, image_counts, color=colors)
    plt.xlabel('Folder Names')
    plt.ylabel('Number of Images')
    plt.title('Image Counts in Each Folder')
    plt.xticks(rotation=45, ha='right')
    plt.show()

dataset_folder = r'C:\Users\pirat\Desktop\CSE499\unsupervised'
visualize_image_counts(dataset_folder)
```
